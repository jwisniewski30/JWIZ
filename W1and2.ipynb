```python
# Function is used to generate a list of random numbers
# We provide the number of elements to be generated and
# the start and end of range from which the random number to choose.
def createRandomList(num, start, end): 
    # create empty array
    arr = [] 
    tmp = random.randint(start, end) 
    
    
    # for the given elements to generate we iterate the for loop
    for x in range(num): 
          
        # this is to check if the number generated is a duplicate 
        # you can remove this if you want duplicate to exist in array
        while tmp in arr: 
            tmp = random.randint(start, end) 
              
        # add the number to array
        arr.append(tmp) 
      
    return arr 

# Driver's code 
# call the function to generated 10 random numbers between 100 and 200
newlist = createRandomList(10, 100, 200)
# printed the generated numbers using the for loop
print("The Generated Random List is: ")
for num in newlist:
    print(num, end =" ")

# this is just for creating a new line to format the print statements    
print()

# than use the sort method present in the Array call of python
# this function converts the original array to sorted array
# so the newList elements will get sorted
newlist.sort()

# printed the sorted array 
print("The Sorted List is: ")
for num in newlist:
    print(num, end =" ")
```

```python
import math
def convert_to_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number += (number_stack.pop() * (math.pow(10, i)))
    return final_number
```

```python
for number_tuple in permutations(range(3)):
    number_stack = list(dropwhile(lambda x: x <= 0, number_tuple))
    print(convert_to_number(number_stack))
```

```python
from itertools import zip_longest
import csv
```

```python
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
with open("sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        # Here we loop over the first 10 lines in order to not to make the output too big
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```

```python
# Read Candy File
df=pd.read_csv("Desktop\candyhierarchy2017.csv", encoding = "ISO-8859-1")
pd.set_option('display.max_columns', None)
print (df.head())
```

```python
# Find out number of NaN values each column has
df.isna().sum()
```

```python
# Drop multiple columns with more than 70% values missing
df=df.drop(["Internal ID","Click Coordinates (x, y)","Unnamed: 113", "Q7: JOY OTHER", "Q8: DESPAIR OTHER", "Q9: OTHER COMMENTS"], axis=1)
```

```python
df.shape
```

```python
df.rename(columns={
    'Q1: GOING OUT?': 'Going Out?',
'Q2: GENDER' : 'Gender',
'Q3: AGE' : 'Age',
'Q4: COUNTRY': 'Country',
'Q5: STATE, PROVINCE, COUNTY, ETC' : 'Location',
'Q10: DRESS': 'Dress' ,
'Q11: DAY' : 'Day',
'Q12: MEDIA [Daily Dish]' : 'Daily Dish',
'Q12: MEDIA [Science]' : 'Science',
'Q12: MEDIA [ESPN]' : 'ESPN',
'Q12: MEDIA [Yahoo]' : 'Yahoo'
}, inplace=True)
```

```python
df.columns = [x for x in df.columns.str.replace("|",'')]
```

```python
df.columns
```

```python
# Clean Gender Column
print (df['Gender'].isna().value_counts())
```

```python
# 41 records do not have a gender specified
# Replace NaN values in Gender with "Other"
df['Gender'].fillna(value="Other", inplace=True)
```

```python
# Combine and merge datasets
# Create age dataframe
age_df = df['Age'].copy()
print(age_df)
```

```python
## Create going out dataframe
attends_df = df['Going Out?'].copy()
print(attends_df)
```

```python
# Joining dataframes
result = pd.concat([age_df, attends_df], axis=1, join='inner')
display(result)
```
